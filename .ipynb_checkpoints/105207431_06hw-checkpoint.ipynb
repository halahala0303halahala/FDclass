```python
import turtle
import math
bob = turtle.Turtle()

def isosceles(t, r, theta):
    y=r*math.sin(theta*math.pi/180)
    
    t.rt(theta)
    t.fd(r)
    t.lt(90+theta)
    t.fd(2*y)
    t.lt(90+theta)
    t.fd(r)
    t.lt(180-theta)

def polypie(t, n, r):
    theta = 360.0/n
    for i in range(n):
        isosceles(t, r, theta/2)
        t.lt(theta)
    turtle.mainloop()
r=float(input("enter the r of polypie:"))
n=int(input("enter the n of polypie"))
polypie(bob , n, r)
```

```python
import turtle
import math
bob = turtle.Turtle()

def arc(t,r,angle):
    arc_length=2*math.pi*r*angle/360
    n=int(arc_length/3)+1
    step_length=arc_length/n
    step_angle=angle/n
    
    for i in range(n):
        t.fd(step_length)
        t.lt(step_angle)

def petal(t,r,angle):
    for i in range(2):
        arc(t,r,angle)
        t.lt(180-angle)#畫完整的花        
def flower(t,n,r,angle):
    for i in range(n):
        petal(t,r,angle)
        t.lt(360.0/n)
    turtle.mainloop()
n=int(input("enter the n of flower"))
r=float(input("enter the r of flower"))
angle=float(input("enter the angle of flower"))
flower(bob,n,r,angle)
```

```python
import turtle
import math
bob = turtle.Turtle()

def draw_spiral(t,n,length,a,b):
    theta=0.0
    for i in range(n):
        t.fd(length)
        dtheta=1/(a+b*theta)
        
        t.lt(dtheta)
        theta += dtheta
    turtle.mainloop()
n=int(input("enter the n of draw_spiral"))
length=float(input("enter the length of draw_spiral"))
a=float(input("enter the a of draw_spiral"))
b=float(input("enter the b of draw_spiral"))
draw_spiral(bob,n,length,a,b)
```

```python
import turtle
import math
bob = turtle.Turtle()

def N(t):
    bob.lt(90)
    bob.fd(50*(3**(0.5)))
    bob.lt(210)
    bob.fd(100)
    bob.lt(150)
    bob.fd(50*(3**(0.5)))
    bob.rt(90)
    #N
def C(t):
    bob.bk(50)
    bob.rt(90)
    bob.fd(50*(3**(0.5)))
    bob.lt(90)
    bob.fd(50)
    #c
def U(t):
    bob.rt(90)
    bob.fd(50*(3**(0.5)))
    bob.lt(90)
    bob.fd(50)
    bob.lt(90)
    bob.fd(50*(3**(0.5)))
    bob.rt(90)
    #U
def L(t):
    bob.rt(90)
    bob.fd(50*(3**(0.5)))
    bob.lt(90)
    bob.fd(50)
    #l
def A(t):
    bob.lt(90)
    bob.fd(50*(3**(0.5)))
    bob.rt(90)
    bob.fd(50)
    bob.rt(90)
    bob.fd(50*(3**(0.5)))
    bob.bk(50)
    bob.rt(90)
    bob.fd(50)
    bob.lt(180)
    #A
def D(t):
    bob.lt(90)
    bob.fd(50*(3**(0.5)))
    bob.rt(90)
    for i in range(180):
        bob.fd(2*math.pi*25*(3**(0.5))*1/360)
        bob.rt(180/180)
    #D
def E(t):
    bob.fd(50)
    bob.lt(90)
    bob.fd(25*(3**(0.5)))
    bob.lt(90)
    bob.fd(50)
    bob.bk(50)
    bob.rt(90)
    bob.fd(25*(3**(0.5)))
    bob.lt(90)
    bob.fd(50)
    #E
def O(t):
    bob.circle(40)
    #O
def M(t):
    bob.lt(90)
    bob.fd(50*(3**(0.5)))
    bob.rt(90)
    bob.fd(25)
    bob.rt(90)
    bob.fd(50*(3**(0.5)))
    bob.bk(50*(3**(0.5)))
    bob.lt(90)
    bob.fd(25)
    bob.rt(90)
    bob.fd(50*(3**(0.5)))
    bob.lt(90)
    #M
def I(t):
    bob.fd(50)
    bob.bk(25)
    bob.lt(90)
    bob.fd(50*(3**(0.5)))
    bob.lt(90)
    bob.fd(25)
    bob.bk(50)
    bob.rt(180)
    #I
def S(t):
    bob.fd(15)
    for i in range(180):
        bob.fd(2*math.pi*12*(3**(0.5))*1/360)
        bob.lt(180/180)
    for i in range(180):
        bob.fd(2*math.pi*12*(3**(0.5))*1/360)
        bob.rt(180/180)
    bob.fd(15)
    #S
N(bob)
bob.pu()
bob.goto(110,50*(3**(0.5)))#移動到指定的座標
bob.pd()
C(bob)
bob.pu()
bob.goto(170,50*(3**(0.5)))
bob.pd()
C(bob)
bob.pu()
bob.goto(180,50*(3**(0.5)))
bob.pd()
U(bob)
bob.pu()
bob.goto(0,-10)
bob.pd()
L(bob)
bob.pu()
bob.goto(60,-(50*(3**(0.5)))-10)
bob.pd()
A(bob)
bob.pu()
bob.goto(120,-(50*(3**(0.5)))-10)
bob.pd()
N(bob)
bob.pu()
bob.goto(180,-(50*(3**(0.5)))-10)
bob.pd()
D(bob)
bob.pu()
bob.goto(-180,-(50*(3**(0.5)))-20)
bob.pd()
E(bob)
bob.pu()
bob.goto(-120,-(50*(3**(0.5)))-20)
bob.pd()
C(bob)
bob.pu()
bob.goto(-70,-(100*(3**(0.5)))-20)
bob.pd()
O(bob)
bob.pu()
bob.goto(-20,-(100*(3**(0.5)))-20)
bob.pd()
N(bob)
bob.pu()
bob.goto(80,-(100*(3**(0.5)))-20)
bob.pd()
O(bob)
bob.pu()
bob.goto(130,-(100*(3**(0.5)))-20)
bob.pd()
M(bob)
bob.pu()
bob.goto(190,-(100*(3**(0.5)))-20)
bob.pd()
I(bob)
bob.pu()
bob.goto(300,-(50*(3**(0.5)))-20)
bob.pd()
C(bob)
bob.pu()
bob.goto(310,-(100*(3**(0.5)))-20)
bob.pd()
S(bob)
#連在一起 中間加bob.pu()才不會有痕跡
turtle.mainloop()








```
