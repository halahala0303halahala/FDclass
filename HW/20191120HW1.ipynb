```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

### 取後兩項作預測

```python
X = x[:,2:]
Y = y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,
                                                   test_size = 0.2,
                                                   random_state = 0)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
plt.scatter(x_test[:,0] , x_test[:,1] , c=y_predict-y_test, cmap = 'Paired') ;
```

```python
x_arr = np.arange(0,8,0.02)
y_arr = np.arange(0,3,0.02)
X0,Y0 = np.meshgrid(x_arr,y_arr)
P = np.c_[X0.ravel() , Y0.ravel()]
z = clf.predict(P)
Z = z.reshape(X0.shape)
plt.contourf(X0, Y0, Z, cmap='Paired', alpha=0.2)
plt.scatter(x_test[:,0] , x_test[:,1] , cmap='Paired' , c=y_test);
```
