```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 從 [Boxrec Rating](https://boxrec.com/en/ratings) 讀取世界排名前50名拳擊手資料

```python
df = pd.read_html('https://boxrec.com/en/ratings')
```

```python
#確認網頁讀取到需要的表格
len(df)
```

```python
df[1]
```

```python
df = df[1]
```

```python
df.head()
```

```python
#整理表格 將不必要的column去除
df.drop("Unnamed: 3",axis=1,inplace=True)
df.drop("last 6",axis=1,inplace=True)
```

```python
df.head()
```

```python
df = df.drop(25)
df = df.drop(26)
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df.index = range(1,51)
```

```python
#確認最後表格型態
df.head()
```

```python
#世界級拳手平均年齡
df["age"].mean()
```

## 分析拳擊手架勢，觀察何種架式占優

```python
df_ort = df[(df.stance  == "orthodox")]
df_sou = df[(df.stance  == "southpaw")]
```

```python
#拳手架式比例
plt.bar(["orthodox","southpaw"],height=[len(df_ort),len(df_sou)]);
```

```python
print(df_ort.mean())
print(df_sou.mean())
print(df_ort.var())
print(df_sou.var()) 
```

## 世界前50名拳手中，以右架為多數，然平均分數為左架較高
